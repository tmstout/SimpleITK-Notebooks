```python
import SimpleITK as sitk
from myshow import myshow, myshow3d
```

```python
img1 = sitk.ReadImage("/Users/tmstout/Dropbox/CT_Images/CT1.nii.gz")
```

```python
myshow3d(img1,zslices=[40,45])
```

```python
otsuIm = sitk.OtsuThreshold(img1, insideValue = 0, outsideValue = 1)
```

```python
myshow3d(otsuIm, zslices = [45])
```

```python
output = otsuIm[:,:,45]
myshow3d(output)
```

```python
connect = sitk.ConnectedComponent(output)
myshow(connect)
```

```python
filter = sitk.RelabelComponentImageFilter()
filter.SetMinimumObjectSize(10000)
take = filter.Execute(connect)
myshow(take)
```

```python
edge = sitk.GradientMagnitude(take)
myshow(edge)
```

```python
rgb = sitk.LabelToRGB(sitk.Cast(sitk.RescaleIntensity(edge), sitk.sitkUInt32))
myshow(rgb)
```

```python
edge2 = sitk.ZeroCrossing(sitk.Cast(sitk.RescaleIntensity(take), sitk.sitkInt32))
myshow(edge2)
```

```python
rgb2 = sitk.LabelToRGB(edge2)
myshow(rgb2)
```

```python
connect2 = sitk.ConnectedComponent(sitk.Cast(sitk.RescaleIntensity(edge), sitk.sitkInt32))
myshow(connect2)
```

```python
filter = sitk.RelabelComponentImageFilter()
filter.SetMinimumObjectSize(2305) #can check it is counting the right thing by changing the number of objects to 2306
remove = filter.Execute(connect2)
```

```python
myshow(remove)
```

```python
rgb3 = sitk.LabelToRGB(remove)
myshow(rgb3)
```

```python
stats = sitk.LabelStatisticsImageFilter()
stats.Execute(connect2,remove)
print stats
```
