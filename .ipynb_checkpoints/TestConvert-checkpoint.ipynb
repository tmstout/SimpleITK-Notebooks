The goal of this project is to find the waste circumference measurement in a CT abdominal image. 

```python
import SimpleITK as sitk
```

```python
from myshow import myshow, myshow3d
```

```python
img1 = sitk.ReadImage("/Users/tmstout/Dropbox/CT_Images/CT1.nii.gz")
```

```python
myshow3d(img1,zslices=[30,40])
```

#### Figure 1. Middle cross section slice of abdominal CT1.nii.gz. 

```python
otsuIm = sitk.OtsuThreshold(img1, insideValue = 0, outsideValue =1)
```

```python
myshow3d(otsuIm)
```

#### Figure 2. Otsu thresholded image of CT1.nii.gz. 

By thresholding the CT image we can start to segment out the large object that we wish to get the perimeter of. Next the objects that are not needed should be removed using mathematical morphology. 

```python
erodeIm = sitk.BinaryErode(otsuIm, 4)
```

```python
myshow3d(erodeIm)
```

#### Figure 3. Eroded image. Removed smaller unnecessary objects in the image.

The smaller unnecessary objects were removed in the above image using the erode mathematical morphology operation. Next the holes in the object need to be closed.  

```python
closeIm = sitk.BinaryClosingByReconstruction(erodeIm, 14)

sitk.RelabelComponent?
```

```python
myshow3d(closeIm)
```

#### Figure 4. Closed image. 

Now that the object has been sucessfully segmented out and made one homoegenous color, the edges of this object can be found and represented as the perimeter of the object. 

```python
img1_Float32 = sitk.Cast(sitk.RescaleIntensity(closeIm), sitk.sitkFloat32)
```

```python
edgeIm = sitk.CannyEdgeDetection(img1_Float32, lowerThreshold=0.0, upperThreshold=1.0)
```

```python
myshow3d(edgeIm)
```

```python
sobelIm = sitk.SobelEdgeDetection(img1_Float32)
```

```python
myshow3d(sobelIm)
```

#### Figure 5. Single pixel edge response image. 

After trying several edge finding filters, the gradient magnitude one worked the best because it returned a single pixel edge response to the circumference of the segmented object. Now the final thing to do is to count the number of connected components of the white object in the above image. That number will be the perimeter. Then we must convert the number of pixels, i.e. the perimeter into the physical waist circumference measurement.
