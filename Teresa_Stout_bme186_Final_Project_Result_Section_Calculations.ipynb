# This section is purely for the WC calclations of the whole date set. 

```python
import SimpleITK as sitk
from myshow import myshow, myshow3d
```

Image CT 1 has already been calculated in the final report as a demo. Its WC value was reported as 780.046875. 

### Image CT2:

```python
img2 = sitk.ReadImage("/Users/tmstout/Dropbox/CT_Images/CT2.nii.gz")
myshow3d(img2,zslices = [40, 41, 42])
```

```python
img2_2D = img2[:,:,46]
```

```python
otsuIm = sitk.OtsuThreshold(img2_2D, insideValue = 0, outsideValue = 1)
```

```python
connectIm = sitk.ConnectedComponent(otsuIm)
```

```python
filter = sitk.RelabelComponentImageFilter()
filter.SetMinimumObjectSize(10000)
abdominalIm = filter.Execute(connectIm)
```

```python
myshow(abdominalIm)
```

```python
edgeIm = sitk.ZeroCrossing(sitk.Cast(sitk.RescaleIntensity(abdominalIm), sitk.sitkInt32))
```

```python
connectIm2 = sitk.ConnectedComponent(sitk.Cast(sitk.RescaleIntensity(edgeIm), sitk.sitkInt32), fullyConnected= True)
```

```python
filter = sitk.RelabelComponentImageFilter()
filter.SetMinimumObjectSize(1000) #can check it is counting the right thing by changing the number of objects to 2306
perimeterIm = filter.Execute(connectIm2)
rgbIm3 = sitk.LabelToRGB(perimeterIm)
myshow(rgbIm3)
```

```python
stats = sitk.LabelStatisticsImageFilter()
stats.Execute(connectIm2,perimeterIm)
print stats
```

```python
print perimeterIm.GetSpacing()
```

```python
spacing = 0.671875
count = 1130
WC = spacing*count
print "Wasit Circumference for CT2.nii is:"
print WC
```
